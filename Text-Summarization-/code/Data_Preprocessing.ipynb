```python
import json

data = []
with open('dialogsum_dev.json') as f:
    for line in f:
        data.append(json.loads(line))     
```

```python
import re
json_object_list = []
for obj in data:
    dialogue = obj["dialogue"]
    
    #sid = cid
    cur_obj = dict()
    cur_obj['cid'] = "conversation_id_" + obj["fname"].split("_")[1]
    cur_obj['sid'] = "summary_id_" + obj["fname"].split("_")[1]
    
    dialogue = dialogue.split('\n')
    utterances = []
    for d in dialogue:
        clean_utterance = re.sub('\#','', d).split(":")
        dialog_dict = dict()
        dialog_dict["speaker_id"] = clean_utterance[0]
        dialog_dict["utterance"] = clean_utterance[1]
        utterances.append(dialog_dict)
        
    cur_obj['utterances'] = utterances
    cur_obj["summary"] = re.sub('\#','', obj["summary"])
    
    json_object_list.append(cur_obj)
    
json_object = json.dumps(json_object_list, indent = 4)

```

```python
# Writing to output.json
with open("dev.json", "w") as outfile:
    outfile.write(json_object)
```
