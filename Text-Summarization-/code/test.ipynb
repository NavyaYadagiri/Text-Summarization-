```python
import json
import pandas as pd
import re
data = []
```

```python
f =  open('/content/drive/MyDrive/CapstoneProject/medical_conversation_summarization-main/data/Original dataset/dialogsum_test.json')
```

```python
for line in f:
    data.append(json.loads(line))
```

```python
print(len(data))
```

```python
counter = 0
df = pd.DataFrame({'dialogues':[]})
for obj in data:
    dialog_list = []
    dialogue = obj["dialogue"]
    dialog = dialogue.split('\n')
    for d in dialog:
        clean_utterance = re.sub('\#','', d).split(":")
        new_person = "["+ clean_utterance[0] + "]:"
        dialog_list.append(new_person + clean_utterance[1])
    joined_dialog = "".join(dialog_list)
    df.loc[len(df.index)] = [joined_dialog]
```

```python
df = pd.DataFrame({'dialogues':[], 'summary': []})
for obj in data:
    dialog_list = []
    dialogue = obj["dialogue"]
    summary = obj["summary1"]
    summary1 = re.sub('\#','', summary)
    dialog = dialogue.split('\n')
    for d in dialog:
        clean_utterance = re.sub('\#','', d).split(":")
        new_person = "["+ clean_utterance[0] + "]:"
        dialog_list.append(new_person + clean_utterance[1])
    joined_dialog = "".join(dialog_list)
    df.loc[len(df.index)] = [joined_dialog,summary1]
```

```python
def fn_inference(bart,text,max_len=256, nbeam=2):
        print("Inside Inference----")
        bart.eval()
        slines=[]
        slines.append(text)
        with torch.no_grad():
            hypotheses_batch = bart.sample(slines, beam=nbeam, lenpen=2.0, max_len_b=max_len, min_len=5, no_repeat_ngram_size=3)
        return hypotheses_batch[0]
```

```python
df1['AI_Gen_Summary'] = df1['dialogues'].apply(lambda x:fn_inference(bartmodel,x))
```

```python
import nltk
nltk.download('punkt')
from nltk.translate.bleu_score import corpus_bleu, sentence_bleu
# import ast
from nltk import word_tokenize
```

```python
def add_bleu(human_given_summary,predicted_transcript):
    pred_word_list=word_tokenize(predicted_transcript)
    actual_word_list=word_tokenize(human_given_summary)
    reference=[actual_word_list]
    candidate=pred_word_list
    score = sentence_bleu(reference, candidate,weights=(0.75, 0.25, 0, 0))
    return score
```

```python
df1['bleu_score'] = df1.apply(lambda x:add_bleu(x['summary'],x['AI_Gen_Summary']),axis=1)
```
